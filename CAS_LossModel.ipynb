```python
CAS Claim Simulator R Package: A Small Example
###################################################################################
### This example shows how to input assumption and run simulation for a single business line#######
### It covers four claim classes: open claim, claim reopenness, IBNR and UPR (future claims) ########
###Users are expected to set assumptions using the API to avoid lengthy coding in R###############
###However, they may be preferred by advanced R users ####################################
###Suggested R Version: R-3.3.3 or above#################################################
#install required packages

#install.packages(c("R2HTML","fitdistrplus","moments","copula","scatterplot3d"))
#install CAS Simulator package from local zip file (cascsim.zip)
#install.packages("http://www.reserveprism.com/Simulator2/cascsim.zip", repos=NULL, method="libcurl")
#load library
#library(cascsim)
import casc_sim
#A list of functions and objects built in the package. A html help file will open.
#help(package="cascsim")
#load sample data
claimdata = data(claimdata)
#Set the name of the business line
lines = ["Auto"]
#Set the subcategory under each line of business, such as normal claim, high claim, jump claim, etc.
types = ["N"]

#####Create the claim object
AutoN = claimtype(line = "Auto", claimType = "N")

####################distribution assumption setting######################################
#set exposure information as an index
AutoN@exposureIndex <- setIndex(new("Index",indexID="I1",tabulate= FALSE,startDate=as.Date("2012-01-01"), annualizedRate = 0)) # level exposure across time
#set monthly frequency distribution assumption
AutoN@frequency <- new("Poisson", p1 =50)
#set severity trend
AutoN@severityIndex <- setIndex(new("Index",indexID="I2",tabulate= FALSE,startDate=as.Date("2012-01-01"), annualizedRate = 0.02)) #assuming a 2% annual inflation
#set severity distribution assumption
AutoN@severity <- new("Lognormal", p1 =2, p2 =3)
#set deductible distribution assumption. Here we set a universal deductible of 10.
AutoN@deductible <- new("Empirical", empirical=matrix(c(0,1,100,100),2,2))
#set limit distribution assumption. Here we set a universal limit of 1e8.
AutoN@limit <- new("Empirical", empirical=matrix(c(0,1,1e8,1e8),2,2))
#set the probability of zero-amount claims (excluding those small claims below deductibles).
AutoN@p0<-0
#set report lag assumption
AutoN@reportLag <- new("Exponential", p1 =0.1)
#set settlement lag assumption
AutoN@settlementLag <- new("Exponential", p1 =0.05)
#set the copula among severity, report lag and settlement lag
AutoN@iCopula <- TRUE #use copula
AutoN@ssrCopula <- new("CopulaObj", type ="normal", dimension = 3, param = c(0.1,0.2,0.1))#A Gaussian Copula
AutoN@ssrCopula@marginal <- c(AutoN@severity,AutoN@settlementLag,AutoN@reportLag) #Set marginal distributions
#Set loss adjustment expense assumption
AutoN@laeDevFac <- new("DevFac",FacID="F1",FacModel= TRUE,fun="linear",paras =c(5,1.5,0.005,1.2,3)) #LAE=5+1.5*development year+0.005*ultimate loss +1.2* o/s ratio + 3*N(0,1)

#####################IBNER assumption setting#########################################
#IBNER loss development factor (from incurred loss to ultimate loss)
AutoN@fIBNER <- new("DevFac",FacID="D1",FacModel= FALSE,meanList =c(1.2,1.15,1.1,1.05,1),volList =c(0,0,0,0,0))
####################Claim Reopenness assumption setting##################################
#Reopen probability
AutoN@reopen <- new("DevFac",FacID="D2",FacModel= FALSE,meanList =c(0.02,0.015,0.01,0.005,0),volList =c(0.003, 0.002, 0.001, 0.001, 0))
#Reopen loss development factor
AutoN@roDevFac <- new("DevFac",FacID="D3",FacModel= FALSE,meanList =c(1.05,1.1,1,1,1),volList =c(0.00589,0.0037,0.00632,0.00815,0))
#Reopen lag assumption. Time lag between last close date and reopen date
AutoN@reopenLag <- new("Exponential", p1 =0.01)
#Resettlement lag assumption. Time lag between reopen date and resettle date
AutoN@resettleLag <- new("Exponential", p1 =0.25)

##########################Set up simulation project######################################
simobj <- new("Simulation", lines=lines, types=types, claimobjs= list(AutoN))
#Set working folder
simobj@workingFolder <- "C:/temp/CAS/demo1" #input your own working folder
#Set the number of simulations to run
simobj@simNo <- 10
#Set the claim classes to run
simobj@iRBNER <-TRUE #Run open claims
simobj@iROPEN <-TRUE #Run claim reopenness
simobj@iIBNR <-TRUE #Run IBNR
simobj@iUPR <-TRUE #Run future claims
#Run simulation and generate individual claim level data
simdata <- claimSimulation(simobj,claimdata, startDate = as.Date("2012-01-01"), evaluationDate = as.Date("2016-12-31"), futureDate = as.Date("2017-12-31"))
#Summarizing generated data to output descriptive statistics
simSummary <- simSummary(simobj,simdata)
#Generating Triangles
simTriangle <- simTriangle(simobj,claimdata,simdata)
#Generating html report that includes results and assumptions
simReport(simobj, simSummary, simTriangle)

```
